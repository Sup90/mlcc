```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense,LSTM,Embedding,Dropout
import numpy as np

from tensorflow.keras.models import Sequential
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
imdb = keras.datasets.imdb

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
```

```python
max=0
idx=0
for i,arr in enumerate(X_train):
    if max<len(arr):
        max=len(arr)
        idx=i
print("max length: ",max ,"index =" ,idx)

min=0
idx=0
for i,arr in enumerate(X_train):
    if min>len(arr):
        min=len(arr)
        idx=i
print("min length: ",min ,"index =" ,idx)
```

```python
word2index=imdb.get_word_indexget_word_index()
```

```python
sorted(word2index.items(),key=lambda kv:kv[1])
```

```python
word2index={k:(v+3) for k , v in word2index.items()}
word2index["<PAD>"]=0
word2index["<START>"]=1
word2index["<UNK>"]=2
word2index["<UNUSED>"]=3
```

```python
sorted(word2index.items(),key=lambda kv:kv[1])
```

```python
index2word=dict([(value,key) for (key,value) in word2index.items()] )
```

```python
index2word
```

```python
" ".join([index2word.get(i,"?") for i in X_train[0]])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
X_train=pad_sequences(X_train,value=word2index["<PAD>"],padding="post",maxlen=256)
X_test=pad_sequences(X_test,value=word2index["<PAD>"],padding="post",maxlen=256)
```

```python
X_train.shape
```

```python
print(X_train[205])
```

```python
model=Sequential()
model.add(Embedding(10000,64))
model.add(LSTM(32,dropout=0.2))
model.add(Dense(1,activation="sigmoid"))

model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=512,validation_data=(X_test,y_test),verbose=1
```
