# 다차원 Array의 경우, length 와 Size가 다름!!!!

```python
x=np.random.randint(1,11,50)
y=x+np.random.randint(1,5,size=x.size)
z=np.column_stack((x,y))
z[1:10]
```

```python
z[0:5]
```

```python
z[:,0]
```

```python
z[:,1]
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(8,4))#1행 2열 형태의 멀티 plot
ax1.scatter(z[:,0],z[:,1],marker=',',c='r')
ax2.hist(z,bins=10)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(8,4))#1행 2열 형태의 멀티 plot
ax1.scatter(z[:,0],z[:,1],marker='^',c='Y',edgecolor='k')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.hist(z,bins=10)
```

```python
delta=0.025
x=np.arange(-3,3,delta)
y=np.arange(-2,2,delta)
xx,yy=np.meshgrid(x,y)
```

```python
z1=np.exp(-xx**2-yy**2)
z2=np.exp(-(xx-1)**2-(yy-1)**2)
z=(z1-z2)*2
```

```python
fig, ax=plt.subplots()
ax.contour(xx,yy,z)
```
