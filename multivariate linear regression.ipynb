```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
import pandas as pd

```

```python
diabetes=load_diabetes()
diabetes.feature_names
```

```python
diabetes.data.shape
```

```python
df=pd.DataFrame(data=diabetes.data,columns=diabetes.feature_names)
```

```python
df['target']=diabetes.target
```

```python
x=df['bmi'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.drop("target",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
diabetes.target.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split( df,diabetes.target, test_size=0.2, random_state=41)
```

```python
regr = linear_model.LinearRegression()
```

```python

# training set   model 

regr.fit(X_train, y_train)
```

```python
diabetes_y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
print('Intercept: \n', regr.intercept_)
```

```python
# MSE(mean squared error) 
print("Mean squared error: {:.2f}"
                        .format(mean_squared_error(y_test, diabetes_y_pred)))
```

```python
# R2  
print("Variance score: {:.2f}".format(r2_score(y_test, diabetes_y_pred)))
```
