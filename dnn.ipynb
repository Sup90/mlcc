```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, Activation,Dropout
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_boston
from sklearn.preprocessing import MinMaxScaler
#편향되있을때 Minmax
#정규분포할때 standard
```

```python
boston=load_boston()
```

```python
df=pd.DataFrame(data=boston.data,columns=boston.feature_names)
```

```python
x=boston.data
y=boston.target
```

```python
y[:10]
```

```python
x.shape
```

```python
y.shape
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.2)
```

```python
print(xtr.shape)
print(ytr.shape)
print(xte.shape)
print(yte.shape)
```

```python
sc=MinMaxScaler()
```

```python
xtr=sc.fit_transform(xtr)
xte=sc.transform(xte)
```

```python
model=Sequential()
model.add(Dense(64,input_shape=(13,))) #inputut_shape=(13,)은 이미지의 경우 다차원으로 이루어지기 때문에 (x,)의 x 옆자리를 항상 비워둠

model.add(Activation('relu'))
model.add(Dense(50))
model.add(Activation('relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer='adam',metrics=['mse','mae'])
```

```python
history=model.fit(xtr,ytr,epochs=500,batch_size=32,validation_data=(xte,yte)) #변수에 저장하면 로그가 다 기록되어 활용 가능
```

```python
model.evaluate(xte,yte)
```

```python

y_pred=model.predict(xte)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(yte,y_pred)
```

```python
r2_score(yte,y_pred)
```

```python
plt.scatter(yte,y_pred)
plt.plot([yte.min(),yte.max()],[y_pred.min(),y_pred.max()],'--',c='r')
```

```python
plt.plot(history.history['mean_absolute_error'],label='test')
plt.plot(history.history['val_mean_absolute_error'],label='validation')
```
