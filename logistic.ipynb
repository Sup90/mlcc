```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
dataset=pd.read_csv("C:/Users/COM/Desktop/python class/machineLearningBasic-master/machineLearningBasic-master/datasets/Social_Network_Ads.csv")
```

```python
df=dataset
```

```python
df.head()
```

```python
df['Purchased'].value_counts()
```

```python
x=df.iloc[:,[2,3]].values.astype("float32")
y=df.iloc[:,4].values.astype("float32")
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.2)
```

```python
sc=StandardScaler()
```

```python
xtr
```

```python
xtr_sc=sc.fit_transform(xtr)
xte_sc=sc.transform(xte)
#fit을 두번하면 스케일 기준이 변경되어 fit은 한번만
lr_clf=LogisticRegression()
lr_clf.fit(xtr_sc,ytr)
```

```python
y_pred=lr_clf.predict(xte_sc)
```

```python
y_pred_pro=lr_clf.predict_proba(xte_sc)
```

```python
y_pred
```

```python
a=pd.DataFrame(y_pred_pro)
```

```python
a['label']=y_pred
```

```python
y_pred_pro>0.6  #임계값 조정 디폴트는 0.5
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(yte,y_pred,labels=[1,0]))
```
